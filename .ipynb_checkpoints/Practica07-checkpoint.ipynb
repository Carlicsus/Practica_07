```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine(f"mysql+pymysql://root:1234@localhost:3307/hospitudo")
```

```python
df_usuarios = pd.read_sql("SELECT * FROM tbb_usuarios", engine)
df_usuarios.head()  
```

```python
df_usuarios.size
```

```python
print(len(df_usuarios))
```

```python
df_usuarios.info()
```

```python
df_usuarios = df_usuarios.dropna(subset=['fecha_registro'])
```

```python
print(len(df_usuarios))
```

```python
df_usuarios = df_usuarios[['nombre_usuario', 'contrasena', 'estatus', 'numero_telefonico_movil']]
```

```python
print(len(df_usuarios))
df_usuarios
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    conn.execute(text("""
        CREATE TEMPORARY TABLE tmp_usuarios_filtradas (
            nombre_usuario VARCHAR(60),
            contrasena VARCHAR(60),
            estatus ENUM('Activo', 'Inactivo', 'Bloqueado', 'Suspendido'),
            numero_telefonico_movil CHAR(19)
        );
    """))
```

```python
df_usuarios.to_sql(
    name='tmp_usuarios_filtradas',
    con=engine,
    if_exists='append',  # usamos append porque la tabla ya existe
    index=False
)
```

```python
verification_query = f"SELECT * FROM tmp_usuarios_filtradas LIMIT 5"
result = pd.read_sql(verification_query, engine)
print(result)
```

### Ejercicio 2: Datos de API

Importar datos de una API 

```python
import requests
import pandas as pd

url = "https://swapi.py4e.com/api/people"
all_people = []

while url:                     
    data = requests.get(url).json()
    all_people.extend(data["results"])
    url = data["next"]        
    
def get_planet_name(url):
    return requests.get(url).json().get("name", "Desconocido")
    
df_api = pd.json_normalize(all_people)
df_api['homeworld_name'] = df_api['homeworld'].apply(get_planet_name)
```

```python
df_api
```

```python
df_api.dropna()
```

```python
df_api = df_api[['name', 'birth_year', 'gender', 'homeworld_name']]
```

```python
df_api
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    conn.execute(text("""
        CREATE TEMPORARY TABLE tmp_sw_personajes (
            name VARCHAR(100),
            birth_year VARCHAR(20),
            gender VARCHAR(20),
            homeworld_name VARCHAR(100)
        );
    """))
```

```python
df_api.to_sql(
    name='tmp_sw_personajes',
    con=engine,
    if_exists='append', 
    index=False
)
```

```python
verification_query = f"SELECT * FROM tmp_sw_personajes LIMIT 35"
result = pd.read_sql(verification_query, engine)
print(result)
```
